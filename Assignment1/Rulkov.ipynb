## Rulkov map

```python
def funX(x,y,alpha):
    x_n1=0
    if x <= 0:
        x_n1=(alpha/(1-x))+y
    elif x > 0 and x < alpha+y:
        x_n1=alpha+y
    elif x >= alpha+y:
        x_n1=-1
    return x_n1
        
```

```python
def funY(x,y,mu,sigma):
    return (y-mu*(x+1)+mu*sigma)
```

## Function to generate spike

```python
import matplotlib.pyplot as plt
def generate_spike(alpha,mu,sigma,title):
    x_axis=range(10000)
    l=[]
    x=0
    y=0
    for i in range(10000):
        x_t=funX(x,y,alpha)
        x=x_t
        l.append(x)
        y_t=funY(x,y,mu,sigma)
        y=y_t
    plt.plot(x_axis, l, label = "spike") 
    plt.xlabel('n') 
    plt.ylabel('x') 
    plt.title(title)
    plt.show()
    
    
    
```

## a) Periods of Silence Plot

```python
title="periods of silence"
alpha=4
mu=0.001
sigma=-0.01
generate_spike(alpha,mu,sigma,title)
```

## b) Tonic spiking plot

```python
title="tonic spiking"
alpha=4
mu=0.001
sigma=0.01
generate_spike(alpha,mu,sigma,title)
sigma=0.1
generate_spike(alpha,mu,sigma,title)
```

## c) Bursts of spikes plot

```python
title="bursts of spikes"
alpha=6
mu=0.001
sigma=-0.01
generate_spike(alpha,mu,sigma,title)
sigma=0.386
generate_spike(alpha,mu,sigma,title)
```
